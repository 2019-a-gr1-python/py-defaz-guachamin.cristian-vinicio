## Análisis benchmarking

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df_HDD = pd.read_csv('E:/00_python/scrapy/HDD_Benchmarks.csv')
```

```python
df_HDD.head()
```

### 1)  10 discos HDD que lideran el ranking

```python
df_1 = df_HDD[['nombre_hdd','disco_rank']]
df_10_ranking = df_1.sort_values(by='disco_rank',ascending = False).tail(10)
df_10_ranking.plot.barh(x='nombre_hdd', y='disco_rank')
plt.show()
```

### 2) 10 discos HDD con mejor raiting

```python
df_2 = df_HDD[['nombre_hdd','disco_rating']]
df_10_raiting = df_2.sort_values(by='disco_rating',ascending = False).head(10)
df_10_raiting.plot.barh(x='nombre_hdd', y='disco_rating')
plt.show()
```

### 3) 20 discos más caros (En dólares)

```python
df_3 = df_HDD[['nombre_hdd','price']]
df_10_caros = df_3.sort_values(by='price',ascending = False).tail(10)
df_10_caros.plot.barh(x='nombre_hdd', y='price')
plt.show()
```

### 4) Discos de Intel en el ranking top 5

```python
df_4 = df_HDD[df_HDD['nombre_hdd'].str.contains("Intel")]
df_Intel_top10 = df_4[['nombre_hdd','disco_rank']].sort_values(by='disco_rank',ascending = False).tail(5)
df_Intel_top10.plot( x='nombre_hdd', y='disco_rank' ,figsize=(14,4))
plt.show()
```

### 5) 5 Discos de Gygabyte mejor puntuados

```python
df_5 = df_HDD[df_HDD['nombre_hdd'].str.contains("Gigabyte")]
df_gyga_top5 = df_5[['nombre_hdd','disco_rating']].sort_values(by='disco_rating',ascending = False).head(5)
df_gyga_top5.plot( x='nombre_hdd', y='disco_rating' ,figsize=(14,4))
plt.show()
```

### 6) Los 10 discos más costosos de SanDisk

### 11)  Procesadores AMD mejor rankeados

```python
df_CPU = pd.read_csv('E:/00_python/scrapy/CPU_Benchmarks.csv')
df_CPU.head()
```

```python
df_11 = df_CPU[df_CPU['cpu_name'].str.contains("AMD")]
df_amd_10 = df_11[['cpu_name','cpu_rank']].sort_values(by='cpu_rank',ascending = False).head(5)
df_amd_10.plot.barh(x='cpu_name', y='cpu_rank' ,figsize=(10,4))
plt.show()
```

### 12) Número de procesadores por fabricante en el top 100 

```python
df_12 = df_CPU[['cpu_name','cpu_rank']]
df_100_ranking = df_12.sort_values(by='cpu_rank',ascending = False).head(100)
num_AMD = len(df_100_ranking[df_100_ranking['cpu_name'].str.contains("AMD")].index)
num_Intel = len(df_100_ranking[df_100_ranking['cpu_name'].str.contains("Intel")].index)
AMD = ['AMD', num_AMD]
Intel = ['Intel', num_Intel]
Otros = ['Otros', 100-num_AMD-num_Intel]

df_resultados = pd.DataFrame([AMD, Intel, Otros], columns=['Fabricante','Num'])
df_resultados.plot(kind='pie', y = 'Num', autopct='%2.1f%%', figsize=(5,5),
startangle=45, shadow=False, labels=df_resultados['Fabricante'], fontsize=10)
plt.show()
```
