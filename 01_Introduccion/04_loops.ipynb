```python
for numero in range(10):
    print(f"Número: {numero}")
```

```python
for numero in range(10):
    if(numero == 4 or numero == 8):
        continue
    print(f"Indice: {numero}")
    # No se ejecutará ni el 4 ni el 8
```

```python
for numero in range(10):
    if(numero == 4 or numero == 8):
        break
    print(f"Indice: {numero}")
```

```python
numeroAuxiliar = 0
while numero < 10:
    print(f"Numero auxiliar: {numeroAuxiliar}")
```

```python
numeroAuxiliar = 0
while numeroAuxiliar < 10:
    print(f"Numero auxiliar: {numeroAuxiliar}")
    numeroAuxiliar += 1
```
